```python
total_cases = covid_df.new_cases.sum()
```

```python
total_deaths = covid_df.new_deaths.sum()
total_deaths
```

```python
#ration of reported death dived by total cases
Ratio = total_deaths/total_cases
Ratio
```

```python
print('the number of reported cases is{} and the number of reported deaths {}'.format(int(total_cases ), int(total_deaths)))
print('the overall reported death case is{:.2f}%,'.format(Ratio*100))
```

```python
#wht is overall numner of test conducted if a total of 935310 tests were conducted before daily test numbers were reported
intial_test=935310
total_tests = covid_df.new_tests.sum()+ intial_test
total_tests
```

```python
print('the overall reported test case is {}'. format(int(total_tests)))
```
